**MERGING THE CSV FILES**

*We use the date in the df2 to merge the two tables, we don't need the time stamp.*

```python
# Convert dates using dayfirst parameter
df1['date'] = pd.to_datetime(df1['date'])
df2['date'] = pd.to_datetime(df2['Timestamp IST'], dayfirst=True).dt.normalize()

# Merge
df3 = pd.merge(
    df2,
    df1[['date', 'value', 'classification']],
    on='date',
    how='left'
)
df3.head()
```

**DROPPING UNWANTED COLUMNS**

```python
# Drop multiple columns at once
df3.drop(columns=[
    'Account','Coin','Timestamp IST','Start Position',
    'Transaction Hash','Order ID','Trade ID','Crossed','Timestamp'
], inplace=True)
```

```python
df3 = df3.dropna()
```

## EXPORTING GRAPHS

```python
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Ensure outputs folder exists
os.makedirs("outputs", exist_ok=True)

# 1. Bar Plot: Avg Closed PnL for each sentiment
plt.figure(figsize=(8,5))
sns.barplot(x='classification', y='Closed PnL', data=df3, errorbar=None)
plt.title('Average Closed PnL by Sentiment')
plt.ylabel('Average Closed PnL')
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig('outputs/avg_pnl_by_sentiment.png')
plt.close()

# 2. Bar Plot: Avg Trade Volume (Size USD) for each sentiment
plt.figure(figsize=(8,5))
sns.barplot(x='classification', y='Size USD', data=df3, errorbar=None)
plt.title('Average Trade Volume (USD) by Sentiment')
plt.ylabel('Average Size USD')
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig('outputs/avg_volume_by_sentiment.png')
plt.close()

# 3. Stacked Bar: Buy vs Sell counts in each sentiment
trade_counts = df3.groupby(['classification', 'Side']).size().unstack(fill_value=0)
trade_counts.plot(kind='bar', stacked=True, figsize=(8,5))
plt.title('Trade Direction Counts by Sentiment')
plt.ylabel('Number of Trades')
plt.xticks(rotation=30)
plt.tight_layout()
plt.savefig('outputs/trade_direction_by_sentiment.png')
plt.close()

# 4. Line Chart: Sentiment Score vs Average Daily PnL
daily_pnl = df3.groupby(['date', 'value'])['Closed PnL'].mean().reset_index()
plt.figure(figsize=(10,5))
sns.lineplot(x='value', y='Closed PnL', data=daily_pnl, marker='o')
plt.title('Sentiment Score vs Average Daily PnL')
plt.xlabel('Sentiment Score (Fear=0  Greed=100)')
plt.ylabel('Average Daily Closed PnL')
plt.tight_layout()
plt.savefig('outputs/sentiment_score_vs_pnl.png')
plt.close()

print(" All plots saved in 'outputs/' folder.")
```
